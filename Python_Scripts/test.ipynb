```python
import pandas as pd
df = pd.read_csv("/lunarc/nobackup/projects/lu2024-17-19/J_rAAven/BRAVE/0_data/final_fragments_summary.csv")
df
```

```python
import numpy as np
# create a new colum Barcode count where you count the number of barcodes in each row
df['BC_count'] = df['BC'].str.split(',').apply(len)
df['BC_adjusted_count'] = np.log2(df['BC_count'] * df['Normalized_RNAcount'])
```

```python
df.head()
```

```python
# get the top 20 fragmentsb based on the Normalized RNA count for eveery group in the data
df_top_25 = pd.DataFrame()
for group in df['Group'].unique():
    df_group = df[df['Group'] == group]
    df_group = df_group.sort_values(by='log2NormCount', ascending=False)
    df_top_25 = pd.concat([df_top_25, df_group.head(25)])
df_top_25.head(5)
```

```python
# Create a bar plot of the top 25 fragments for each group where the x-axis is the Genename of the fragment and the y-axis is the log2NormCount
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_counts(df: pd.DataFrame, top_n: int, group_col: str=None):
    if group_col is not None:
        df = df[df['Group'] == group_col]
    
    df_top_n = df.sort_values(by='BC_adjusted_count', ascending=False).head(top_n)
    
    plt.figure(figsize=(10, 5))
    if group_col is not None:
            sns.barplot(x='LUTnr', y='BC_adjusted_count', data=df_top_n)
    else:
        sns.barplot(x='LUTnr', y='BC_adjusted_count', hue='Group', data=df_top_n)
    plt.xticks(rotation=45, ha='right', fontsize=6)
    plt.title(f'Top {top_n} fragments in {group_col}')
    plt.show()
```

```python
plot_top_counts(df, 50, 'DNA_AAVlib_DNAse_30cpc')
```
