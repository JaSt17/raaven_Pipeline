```python
# read in two text files and store every line in a set
with open("/lunarc/nobackup/projects/lu2024-17-19/J_rAAven/BRAVE/0_data/SortedFragments_all.txt") as file1, open("/lunarc/nobackup/projects/lu2024-17-19/J_rAAven/BRAVE/0_data/SortedFragments_all_python.txt") as file2:
    lines1 = file1.readlines()
    lines2 = file2.readlines()
    fragments1 = set(lines1)
    fragments2 = set(lines2)

# get the number of intersections between the two sets
intersection = fragments1.intersection(fragments2)
print(len(intersection))

# get the number of differences between the two sets
difference = fragments2.difference(fragments1)
print(len(difference))
for i in difference:
    print(i)

# for some reason i get 240 more sequences with the python pipeline than with the bash pipeline
```
