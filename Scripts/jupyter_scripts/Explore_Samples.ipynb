```python
import pandas as pd
import numpy as np
```

```python
# Load the data
allSamples = pd.read_csv('/lunarc/nobackup/projects/lu2024-17-19/J_rAAven/BRAVE/0_data/allSamplesDataTable.csv')
# sort by LUTnrs
allSamples
```

```python
# get the number of samples for each group without filtering out duplicated LUTnrs
groupcounts = allSamples.groupby('Group').size()
# sum all groupcounts except the ones that start with D or includes 'All'
sumgroupcounts = groupcounts[~groupcounts.index.str.contains('D|All')].sum()
print(groupcounts)
```

```python
# Show the table without duplicates
allSamples1 = allSamples.drop_duplicates(subset=['Group', 'LUTnrs', 'structure'])
allSamples1.groupby('Group').size()
```

Here we are combining groups into groups and renaming them before checking for duplicates

```python
# Combine the two groups into 'DNA_AAVlib_DNAse' into one group 'DNA_resistant_AAV' and rename 'DNA_pscAAVlib' to 'Plasmid_library'
allSamples.loc[:,'Group'] = allSamples.loc[:,'Group'].replace({
    'DNA_pscAAVlib': 'Plasmid_library',
    'DNA_AAVlib_DNAse_30cpc': 'DNA_resistant_AAV',
    'DNA_AAVlib_DNAse_3cpc': 'DNA_resistant_AAV'
})


# Get a list of all groups
groups = list(allSamples['Group'].unique())
# collect all mRNA in vivo samples into one group
# Exclude groups containing 'DNA', 'pNeuron', 'HEK293', 'All', 'Trsp', 'Plasmid'
exclude_terms = ['DNA', 'pNeuron', 'HEK293', 'All', 'Trsp', 'Plasmid']
groups1 = [group for group in groups if all(term not in group for term in exclude_terms)]
# Combine all groups containing 'infectiv_AAV' into one group
for group in groups1:
    allSamples.loc[:,'Group'] = allSamples.loc[:,'Group'].replace(group, 'invivo_infectiv_AAV')
# combine all mRNA samples into one group
exclude_terms = ['DNA', 'Plasmid']
groups2 = [group for group in groups if all(term not in group for term in exclude_terms)]
for group in groups2:
    allSamples.loc[:,'Group'] = allSamples.loc[:,'Group'].replace(group, 'all_infective_AAV')
```

There are cases where we have the same LUTnr multiple times in the data since the sequences were the same in ortholog sequences that we used to generate the fragments.

```python
# filter so we have every LUTnr only once for every group
allSamples = allSamples.drop_duplicates(subset=['Group', 'LUTnrs', 'structure'])
allSamples.groupby('Group').size()
```

```python
# remove all samples that are in the group DNA_pscAAVlib_Prep2 or in all_infective_AAV
allSamples = allSamples[~allSamples['Group'].str.contains('DNA_pscAAVlib_Prep2|all_infective_AAV')]
```

```python
# Save the csv file with the new groups
allSamples.to_csv('/lunarc/nobackup/projects/lu2024-17-19/J_rAAven/BRAVE/0_data/allSamplesDataTable_adjust_groups.csv', index=False)
```
